# Problem 1: Student Grades Analyzer

## Write a function analyze_grades(*grades) that:

* ## Calculates the average grade.

* ## Uses match-case to return:

* * ## "Excellent" if avg ≥ 90

* * ## "Good" if avg ≥ 75

* * ## "Pass" if avg ≥ 50

* * ## "Fail" otherwise
  * analyze_grades(90, 80, 100) → "Excellent"


```python
def analyze_grades(*grades):
    if not grades:
        return "No grades"
        
    avg = sum(grades)/len(grades)
    
    match grades:
        case _ if avg >= 90:
            return "Excellent"
        case _ if avg >= 75:
            return "Good"
        case _ if avg >= 50:
            return "Pass"
        case _:
            return "Fail"

print(analyze_grades(90, 80, 100))
    
```

# Problem 2: Flexible Calculator

## Write a function calculator(operation, *args, precision=2) that:

* ## Supports "sum", "multiply", "max", "min", "average".

* ## Uses a loop (for) to compute results.

* ## Uses match-case to decide the operation.

* ## Round to precision decimals.

* calculator("average", 10, 20, 30) → 20.0


```python
def calculat(operation, *args, precision=2):
    match operation:
        case "sum":
            for i in args:
                result = sum(args)
        case "multiply":
            for i in args:
                result = i*args 
        case "max":
            for i in args:
                result = max(args)
        case "min":
            for i in args:
                result = min(args)

        case "average":
            for i in args:
                result = sum(args)/len(args)
                
    return round(result, precision)
print(calculat("average",10, 20, 30))
```

# Problem 3: Word Formatter with Conditions

## Write a function format_words(words, **options) where:

* ## words is a list of strings.

* ## If upper=True, convert all to uppercase.

* ## If reverse=True, reverse each word.

* ## If repeat=n, repeat each word n times.

* ## Return the final list.

* format_words(["ai", "python"], upper=True, repeat=2)
* → ["AI AI", "PYTHON PYTHON"]


```python
def format_word(words, **options):
    lst=[]
    for word in words:
        if options.get("upper"):
            word= word.upper()
        if options.get("reverse"):
            word=word.reverse()
        if "repeat" in options:
            n= options["repeat"]
            word= (word+" ")*n
        lst.append(word)
    return lst
print(format_word(["ai","python"], upper=True, repeat = 2))
```

# Problem 4: Prime Number Generator

* ## Write a function generate_primes(n) that uses a while loop to return all prime numbers up to n.

* generate_primes(10) → [2, 3, 5, 7]


```python
def generate_primes(n):
    lst=[]
    num = 2
    while num <= n :
        is_prime= True
        div = 2
        while div * div <= num :
            if num % div == 0 :
                is_prime = False
                break
            div += 1
        if is_prime:
            lst.append(num)
        num += 1
    return lst
print(generate_primes(10))
        
```

# Problem 5: Employee Salary Manager

## Write a function employee_salary(basic_salary, *bonuses, tax=0.1, **deductions) that: 

* ## Adds all bonuses to the basic_salary using a loop.

* ## Subtracts all deductions (sum of kwargs values).
deductions = الخصومات كما في المثال القادم
 
    insurance: 500,
    loan: 200,
    retirement: 100


* ## Applies tax on final salary.

* ## Uses conditions: if salary < 3000 → "Low", else "High".

* employee_salary(5000, 200, 300, tax=0.15, insurance=500, loan=200, retirement = 100)


```python
def employee_salary(basic_salary, *bonuses, tax=0.1, **deductions):
    total_salary= basic_salary
    for bonus in bonuses:
        total_salary += bonus

    for value in deductions.values():
        total_salary -+ value

    total_salary -= total_salary * tax

    salary_level = "Low" if total_salary < 3000 else "High"

    return total_salary, salary_level

final_salary, final_level = employee_salary(500,200, 300, tax= 0.15, insurance= 500, loan= 200, retirement=100)
print(f"Final Salary: {final_salary} ---------- Level: {final_level}")
        
```

# Problem 6: Math Game with Switch

## Write a function math_game(choice, *numbers) that:

## Uses match-case where:

* ## 1: return sum of even numbers

* ## 2: return sum of odd numbers

* ## 3: return product of numbers divisible by 3

* ## _: return "Invalid choice"

* ## Use loops + conditions to calculate.

math_game(1, 1, 2, 3, 4, 5, 6) → 12   (2+4+6)


```python
def math_game(choice, *number):
    match choice:
        case 1:
            result = 0
            for i in number:
                if i % 2 == 0 :
                    result += i
            return result
        case 2:
            result = 0
            for i in number:
                if i % 2 != 0:
                    result += i
            return result

        case 3 :
            result = 1
            found = False
            for i in number:
                if i % 3 == 0:
                    result *= i
            return result if found else 0

        case _:
            return "Invalid Choice"

print(math_game(1,1,2,3,4,5,6))
```

### ** Problem 8: Student Study Tracker**

**Problem:**

1. Make a class called `Student`.
2. Each student has:

   * `name`
   * `year` (e.g., 1, 2, 3 … for year of study)
   * `weekly_hours` (list of study hours per day for a week)
3. All students share:

   * `school_name = "Future Academy"`
   * `campus = "Downtown"`
4. Write a method `total_hours()` that calculates the total weekly study hours.
5. Write a method `school_info()` that prints the school name and campus.
6. Write a method `check_overload(avg_daily_hours)` that:

   * If `avg_daily_hours > 6`, print `"Overload"`

**Data Example:**

* Student 1: `"Ahmed"`, year 2, weekly_hours = `[2, 5, 4, 6, 7, 3, 5]`



```python
# Define The Student class To Manage Study Data
class Student:
    # Class attributes : shared by all students
    school_name = "Future Academy"
    campus = "Downtown"
    # constructor
    def __init__(self, name, year, weekly_hours):
        self.name= name
        self.year= year
        self.weekly_hours= weekly_hours
        
    # Calculate total study hours
    def total_hours(self):
        return sum(self.weekly_hours)

    def school_info(self):
        print(f"School: {self.school_name}, campus: {self.campus}")

    def check_overload(self):
        # Calculate daily average:
        avg_daily_hours=sum(self.weekly_hours)/len(self.weekly_hours)
        if avg_daily_hours > 6:
            print("OverLoad")
        else:
             print("Normal Workload")

#-----------Testing----------

weekly_hours=[2,5,4,6,7,3,5]
stu=Student("Ahmed", 2, weekly_hours)
stu.school_info()
print(f"Total Weekly hours : {stu.total_hours()}")
stu.check_overload()
```
