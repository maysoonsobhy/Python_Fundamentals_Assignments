6. Write a program for rock-paper-scissors game using match-case.
* rock vs scissors → rock wins
* scissors vs paper → scissors wins
* paper vs rock → paper wins
* same choice → draw

````python
player=input("Enter rock, paper, scissors: ").lower()
computer=input("Enter rock, paper, scissors: ").lower()
match(player, computer):
    case (player, computer) if player == computer:
        print("Draw")
    case ("rock","scissors"):
        print("rock wins")
    case ("scissors", "paper"):
        print("scissors wins")
    case ("paper", "rock"):
        print("paper wins")
    case _:
        print("Computer wins!")
    
````


**write a program that:**



1. Classifies each customer’s experience,Then add a key in each dictionay in the list called Experience and the value is Excellent, Good or Bad:

    * `"Good"` if rating ≥ 3 **and** waiting_time ≤ 10
    * `"Bad"` if rating < 3 **or** waiting_time > 10


2. Determine whether the store performance is “acceptable”:**

  or

````python
data = [{"name":"Sara", "waiting_time":3, "rating":5},
        {"name":"Ali", "waiting_time":15, "rating":3},
        {"name":"Mona", "waiting_time":7, "rating":3},
        {"name":"Omar", "waiting_time":1, "rating":4}]
count = 0
for customer in data:
    current_rating = customer["rating"]
    current_wating = customer["waiting_time"]
    if current_rating >= 4 and current_wating <= 5 :
        customer["Experience"] = "Excellent"
    elif current_rating >= 3 and current_wating <= 10 :
        customer["Experience"] = "Good"
    else:
        customer["Experience"] = "Bad"
    if customer["Experience"] == "Excellent" or customer["Experience"] == "Good":
        count += 1
lenn_data = len(data)
if count > lenn_data / 2:  
        print(f"Result: {count} Store performance acceptable , Store performance needs improvement")
else:
    print(f"Result: {count} The stoe needs to reduce wating time ,Most customers are satisfied overall")
    
````

````python
products = [{"Product_name": "Laptop Bag", 
             "Purchases":120, "reviews": [{"Customer_name":"Sara", "rating":5, "delivery_time":2},
                                          {"Customer_name":"Ali", "rating":3, "delivery_time":7},
                                          {"Customer_name":"Omar", "rating":4, "delivery_time":3},
                                         {"Customer_name":"Nora", "rating":2, "delivery_time":12},
                                         {"Customer_name":"Hana", "rating":5, "delivery_time":1}]},
           {"Product_name": "Headphones", 
            "Purchases":85, "reviews":[{"Customer_name":"Mona", "rating":2, "delivery_time":10},
                                       {"Customer_name":"Yara", "rating":4, "delivery_time":5},
                                      {"Customer_name":"Tamer", "rating":3, "delivery_time":6},
                                      {"Customer_name":"Adel", "rating":1, "delivery_time":15}]},
           {"Product_name": "Smart Watch", 
           "Purchases":200,"reviews":[{"customer_name":"Lina", "rating":5, "delivery_time":2},
                                     {"Customer_name":"Ola", "rating":4, "delivery_time":3},
                                     {"Customer_name":"Farha", "rating":5, "delivery_time":1},
                                     {"Customer_name":"Rami", "rating":3, "delivery_time":7},
                                     {"Customer_name":"Hadir", "rating":4, "delivery_time":4}]}]
top_purchases = 0
top_purchases_name = ""
top_excellent = 0
top_excellent_name = ""
top_avg_delivery = 0
slowest_delivery = ""

#Loop through each product to analyse reviews

for product in products:
    excellent_count = 0
    good_count = 0
    total_time = 0
    
#Calculate experience based on rating and delivery time 
    
    for review in product["reviews"]:
        if review["rating"] >= 4 and review["delivery_time"] <= 3:
            excellent_count += 1
            review["Experience"] = "Excellent"
        elif review["rating"] >= 3 and review["delivery_time"] <= 7:
            good_count += 1
            review["Experience"] = "Good"
        else:
            review["Experience"] = "Bad"

# Calculate product performance            
            
        total_time += review["delivery_time"]
        total_review = len(product["reviews"])
        if good_count+excellent_count >= total_review/2:
            product["Performance"] = "Acceptable"
        else:
            product["Performance"] = "Needs improvement"

# Find the most purchases product
        
        if product["Purchases"] > top_purchases:
            top_purchases = product["Purchases"]
            top_purchases_name = product["Product_name"]
        if excellent_count > top_excellent:
            top_excellent = excellent_count
            top_excellent_name = product["Product_name"]

# Find the slowest dilvery
            
        average_delivery = total_time / total_review
        if average_delivery > top_avg_delivery:
            top_avg_delivery = average_delivery
            slowest_delivery = product["Product_name"]

# Final Results            

for product in products:
    print(f"Product_Name: {product["Product_name"]}")
    print(f"Performance: {product["Performance"]}")
    print("Reviews: ")
    for r in product["reviews"]:
        print(f"Rating: {r["rating"]}, Delivery_time: {r["delivery_time"]}, Experience: {r["Experience"]}")
    print("------------------------------------")
print(f"The most purchased product is : {top_purchases_name} , ({top_purchases})")
print(f"Product of excellent reviews : {top_excellent_name} , ({top_excellent})")
print(f"Product suffering from delivery : {slowest_delivery} , (Average delivery time : {top_avg_delivery})")
        
````
