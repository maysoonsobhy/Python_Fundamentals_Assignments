#### 2.Create a Python program that:

1. Asks the user to enter two numbers.

2. Stores them as variables (make sure to convert them to integers).

3. Performs and prints the results of the following operations:
*  Addition
*  Subtraction
*  Multiplication
*  Division



```python
# Asking and Storing as variables
num1 = int(input("Enter Your First number : "))
num2 = int(input("Enter Your Second number : "))
# Performning and printing results
print(f"The addition result is : {num1 + num2}")
print(f"The subtraction result is : {num1 - num2}")
print(f"The multiplication result is : {num1 * num2}")
print(f"the division result is : {num1 / num2}")
```
