**4. You have two sets:**
* setA = {10, 20, 30, 40, 50}
* setB = {30, 40, 60, 70}

**Your Tasks:**
1. Display the common elements between the two sets (intersection).
2. Show the elements that are only in setA.
3. Combine both sets (union).
4. Add the number 90 to setA.
5. Remove the number 10 from setA.

````python
A = {10,20,30,40,50}
B = {30,40,60,70}
result = A & B
print(f"The intersection : {result}")
result2 = A - B
print (f"The difference : {result2}")
result3 = A | B
print(f"The union : {result3}")
A.add(90)
print(f"Set after adding 90 : {A}")
A.remove(10)
print(f"Set after removing 10 : {A}")
````

Example:

**6. Again for the following grade dictionnary:**

```python
grades = {
    "Ali": 85,
    "Sara": 90,
    "Omar": 78
}
```
1. All students have taken 5 degree bonus, add 5 degrees to the scores to all students
2. Find and print the **name of the student with the highest grade**.

````python
grades = {"Ali":85,"Sara":90,"Omar":78}
grades.update({"Ali" : grades["Ali"]+5,
              "Sara" : grades["Sara"]+5,
              "Omar" : grades["Omar"]+5})
print(f"Dictionary after adding 5 degrees to all students : {grades}")
highest_grade=max(grades,key=grades.get)
print(f"The name with highest grade is : {highest_grade}")
````
